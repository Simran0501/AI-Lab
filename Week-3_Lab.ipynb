# 8 - PUZZLE PROBLEM

```python
from copy import deepcopy


def printMatrix(mat):
    for i in range(3):
        for j in range(3):
            print("%d " % (mat[i][j]), end = " ")
        print()
    print()
initial = []
print("Enter the initial state entries rowwise:")

for i in range(3):          
    row = []
    for j in range(3):      
         row.append(int(input()))
    initial.append(row)
print("\nInitial Board : \n")
printMatrix(initial)
goal = []
print("Enter the goal state entries rowwise:")

for i in range(3):          
    row = []
    for j in range(3):      
         row.append(int(input()))
    goal.append(row)
print("\nTarget Board : \n")
printMatrix(goal)
moves = [[-1, 0],  # go up
         [0, -1],  # go left
         [1, 0],  # go down
         [0, 1]]  # go right

grid = [[0 for col in range(3)] for row in range(3)]


def calcHeuristic(grid, goal):
    count = 0
    for i in range(len(grid)):
        for j in range(len(grid[0])):
            if grid[i][j] != goal[i][j]:
                count += 1

    return count


def isValid(cell):
    if cell[0] >= 0 and cell[0] < len(grid) and cell[1] >= 0 and cell[1] < len(grid[0]):
        return True
    else:
        return False


def puzzleSolver(initial, goal, cost=1):
    opened = []
    visited = []

    h = calcHeuristic(initial, goal)  
    g = 0  

    opened.append([h, g, initial])

    curr = opened.pop()
    visited.append(curr)
    while curr[2] != goal:

        g = curr[1]
        h = curr[0]

        for i in range(len(curr[2])):
            for j in range(len(curr[2][0])):
                
                for a in range(len(moves)):
                    x = i + moves[a][0]
                    y = j + moves[a][1]

                    if isValid([x, y]):
                        if curr[2][x][y] == 0:
                            currCopy = []
                            currCopy = deepcopy(curr[2])
                            currCopy[x][y] = currCopy[i][j]
                            currCopy[i][j] = 0

                            if currCopy not in visited:
                                g2 = g + cost
                                h2 = g2 + calcHeuristic(currCopy, goal)
                                opened.append([h2, g2, currCopy])
                                visited.append(currCopy)
        opened.sort(reverse=True)

        if len(opened) > 0:
            curr = opened.pop()
            print("h = ", curr[0], " g = ", curr[1])
            print("Next state is :\n")
            for row in curr[2]:
                for el in row:
                    print(el, end = ' ')
                print()
            print()


puzzleSolver(initial, goal)
```

# 8-PUZZLE PROBLEM WITHOUT HEURISTIC

```python
import copy
from heapq import heappush, heappop

n = 3

row = [ 1, 0, -1, 0 ]
col = [ 0, -1, 0, 1 ]

# A class for Priority Queue
class priorityQueue:
        def __init__(self):
                self.heap = []
        def push(self, k):
                heappush(self.heap, k)
        def pop(self):
                return heappop(self.heap)
        def empty(self):
                if not self.heap:
                        return True
                else:
                        return False

class node:

        def __init__(self, parent, mat, empty_tile_pos,
                                cost, level):
                self.parent = parent
                self.mat = mat
                self.empty_tile_pos = empty_tile_pos
                self.cost = cost
                self.level = level
        def __lt__(self, nxt):
                return self.cost < nxt.cost

def calculateCost(mat, final) -> int:

        count = 0
        for i in range(n):
                for j in range(n):
                        if ((mat[i][j]) and (mat[i][j] != final[i][j])):
                                count += 1

        return count

def newNode(mat, empty_tile_pos, new_empty_tile_pos,
                        level, parent, final) -> node:
        new_mat = copy.deepcopy(mat)
        print("empty tile pos : ", empty_tile_pos)
        x1 = empty_tile_pos[0]
        y1 = empty_tile_pos[1]
        x2 = new_empty_tile_pos[0]
        y2 = new_empty_tile_pos[1]
        new_mat[x1][y1], new_mat[x2][y2] = new_mat[x2][y2], new_mat[x1][y1]

        cost = calculateCost(new_mat, final)

        new_node = node(parent, new_mat, new_empty_tile_pos,
                                        cost, level)
        return new_node

def printMatrix(mat):

        for i in range(n):
                for j in range(n):
                        print("%d " % (mat[i][j]), end = " ")

                print()

def isSafe(x, y):
        return x >= 0 and x < n and y >= 0 and y < n

def printPath(root):

        if root == None:
                return

        printPath(root.parent)
        printMatrix(root.mat)
        print()

def solve(initial, empty_tile_pos, final):

        pq = priorityQueue()
        cost = calculateCost(initial, final)
        root = node(None, initial,
                                empty_tile_pos, cost, 0)
        pq.push(root)
        while not pq.empty():
                minimum = pq.pop()
                if minimum.cost == 0:
                        printPath(minimum)
                        return
                for i in range(4):
                        new_tile_pos = [
                                minimum.empty_tile_pos[0] + row[i],
                                minimum.empty_tile_pos[1] + col[i], ]

                        if isSafe(new_tile_pos[0], new_tile_pos[1]):
                                child = newNode(minimum.mat,
                                                                minimum.empty_tile_pos,
                                                                new_tile_pos,
                                                                minimum.level + 1,
                                                                minimum, final,)
                                pq.push(child)

initial = []

print("\nEnter Intital State : \n")
for i in range(0, 3):
    temp = []
    for j in range(0, 3):
        el = int(input())
        temp.append(el)
    initial.append(temp)
print("\nIntital State : \n")
printMatrix(initial)

final = []
print("\nEnter Final State : \n")
for i in range(0, 3):
    temp = []
    for j in range(0, 3):
        el = int(input())
        temp.append(el)
    final.append(temp)

print("\nFinal State : \n")
printMatrix(final)

empty_tile_pos = []
print("Enter the coordinate of the empty cell")
x = int(input())
y = int(input())
empty_tile_pos.append(x)
empty_tile_pos.append(y)

solve(initial, empty_tile_pos, final)
```
