# Naive Bayesian Classification

```python
print("Naive Bayesian Classification using Gaussian classifier:")

#importing dataset IRIS

from sklearn.datasets import load_iris
iris_dataset = load_iris()

x_vec = iris_dataset.data
y_vec = iris_dataset.target
print("\nDisplaying data of IRIS\n")
print(x_vec)
print("\n\nDisplaying target value of each row in IRIS dataset\n")
print(y_vec)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_vec, y_vec, test_size = 0.4, random_state = 1)
print("\nTraining set : \n")
for i in range(len(X_train)):
    print(X_train[i] ,  " " , y_train[i] )

print("\nTesting set : \n")
for i in range(len(X_test)):
    print(X_test[i] ,  " " , y_test[i] )

from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)

#Predicting the test set

y_pred = gnb.predict(X_test)
print("Predicted values of Y_vec in testing set :\n")

for i in range(len(X_test)):
    print(y_test[i] ,  " ", y_pred[i])
count = 0
for i in range(len(X_test)):
    if y_pred[i] == y_test[i]:
        count += 1
        
accuracy = (count/len(X_test))
print("\nAccuracy of predicting the flower of IRIS dataset on test set is : ", accuracy*100, "%")
```
