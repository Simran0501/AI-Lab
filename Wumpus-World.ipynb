```python
#-1 wumpus
#9 stench
#8 glitter
#7 pit
#6 gold
#5 breeze

world=[	[0,5,7,5],
		[9,0,8,0],
		[-1,6,7,8],
		[9,0,8,7]	]			#declaration of a world

def getNextposition(agci,agcj):
    print("\nYou are at postion: (" +str(agci) + ","+str(agcj) + ")\n")
    dx = [1,0,-1,0,1]
    for i in range(4):
      x = agci+dx[i]
      y = agcj+dx[i+1]
      if x >= 0 and x < 4 and y >= 0 and y< 4:
        print("You can go to " + str(x)+" "+str(y))
    agni = int(input("\nEnter row position: "))
    agnj = int(input("Enter col position: "))
    print()
    return agni,agnj;


def getPositiondetail(x,y):
  print()
  if world[x][y] == 6:
    print("Congrats! You find the gold. Now get back to initial position!")
    return -10
  if world[x][y] == 9:
    print("You found stench.")
    return 0
  if world[x][y] == 8:
    print("You find a glitter.")
    return 0
  if world[x][y] == 7:
    print("You find a pit.")
    return 0
  if world[x][y] == 5:
    print("You find breeze.")
    return 0
  if world[x][y] == -1:
    print("You are eaten by Wumpus. You lose!")
    return -5
  return 0
  
agi = 0; agj = 0
agci = 0; agcj = 0
found = False
while(1):
    agci, agcj = getNextposition(agi,agj)
    res = int(getPositiondetail(agci,agcj))
    agi = agci
    agj = agcj
    if res == -10:
      found=True
    if res == -5:
      break;
    if agci == 0 and agcj == 0 and found:
      print("You won!")
      break;
```
