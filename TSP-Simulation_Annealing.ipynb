```python
import random as rm
import math

def create_cities(n):
     result = []
     st = set()
     while len(st) < n:
         city = rm.randint(0, n-1)
         st.add(city)
     for el  in st:
         result.append(int(el))
     return result

def find_pathcost(cities, dist_matrix):
     length = len(cities)
     x_coord = rm.randint(0, length-1)
     y_coord = rm.randint(0, length-1)
     while x_coord == y_coord:
         x_coord = rm.randint(0, length-1)
         y_coord = rm.randint(0, length-1)
     cities[x_coord], cities[y_coord] = cities[y_coord], cities[x_coord]
     total_cost = 0
     for i in range(length):
         if dist_matrix[cities[i]][cities[(i+1)%length]] == 0:
             cities[x_coord], cities[y_coord] = cities[y_coord], cities[x_coord]
             return -1
         total_cost += dist_matrix[cities[i]][cities[(i+1)%length]]
     return total_cost

def simulation_annealing(cost, cities, dist_matrix):

     for i in range(100):
         temp = pow(10, 10)
         cooling_factor = 0.99
         end_temp = 0.0001

         while temp > end_temp:
             path_newCost = find_pathcost(cities, dist_matrix)
             if path_newCost == -1:
                 continue
             diff = path_newCost - cost
             if diff < 0 or math.exp((-1 * diff) // temp) > rm.randint(0, 1):
                 cost = path_newCost
             temp *= cooling_factor
     print("Final path with minimum cost : ")
     print(cities)
     return cost



num = int(input("Enter total number of cities : "))
dist_matrix = [[0]*num for _ in range(num)]
for i in range(num):
     for j in range(i+1, num):
         print("Enter distance between ", i+1 , " and ", j+1)
         dist = int(input())
         dist_matrix[i][j] = dist
         dist_matrix[j][i] = dist
print(dist_matrix)
cities = create_cities(num)
cost = find_pathcost(cities, dist_matrix)
result = simulation_annealing(cost, cities, dist_matrix)
print("The final cost of found path is : ", result)


```
